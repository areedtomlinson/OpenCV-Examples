### Opening and Displaying an Image
First we'll import `pyplot` from `matplotlib` and set it up to show images in-line, here in the notebook (instead of trying to open a new window for display).

````python
from matplotlib import pyplot as plt
%matplotlib inline
````

Now we'll use `cv2.imread` to open images, and `plt.imshow` to display them.

### cv2.imread
The first argument is a filename (relative path).

The second (optional) argument specifies how the image should be read:
```
cv2.IMREAD_COLOR : Loads a color image. Any transparency of image will be neglected. It is the default flag.
cv2.IMREAD_GRAYSCALE : Loads image in grayscale mode
cv2.IMREAD_UNCHANGED : Loads image as such including alpha channel
```

`cv2.imread` documentation lives [here](https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_image_display/py_image_display.html).

### Demonstration 1: BGR JPG
A JPG image loaded with the default arguments will have the wrong colorspace to begin with. You have to convert from BGR to RGB colorspace to start out.

The image used here is from NASA, via [Wikimedia Commons](https://commons.wikimedia.org/wiki/File:NASA%27s_Orion_Spacecraft_Parachutes_Tested_at_U.S._Army_Yuma_Proving_Ground.jpg), and is in the public domain.

````python
# Option 1: Single argument
# This appears to default to "cv2.IMREAD_COLOR"
img = cv2.imread('jpg_test.jpg')

# If the image path is incorrect or there was another error, "img" will be None
assert(img is not None)

# Display the image:
plt.imshow(img)
````

````python
# Option 1: Single argument
# This appears to default to "cv2.IMREAD_COLOR"
img = cv2.imread('jpg_test.jpg')

# If the image path is incorrect or there was another error, "img" will be None
assert(img is not None)

# Convert from BGR to RGB, and then display:
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
````

### Demonstration 2: PNG With Transparency
A PNG image with transparency has a 4th channel (the "alpha channel"). If you try to load it as a normal RGB image, you'll get some strange behavior (see below). This type of image has to be loaded with the `cv2.IMREAD_UNCHANGED` argument. 

The image used here is from [Wikimedia Commons](https://commons.wikimedia.org/wiki/File:PNG_transparency_demonstration_2.png), and is licensed under the Creative Commons Attribution-Share Alike 3.0 Unported license.

````python
# Option 1: Single argument
# This appears to default to "cv2.IMREAD_COLOR"
img = cv2.imread('png_transparency_test.png')

# If the image path is incorrect or there was another error, "img" will be None
assert(img is not None)

# Display the image:
plt.imshow(img)
````

````python
img = cv2.imread('png_transparency_test.png', cv2.IMREAD_COLOR)

# If the image path is incorrect or there was another error, "img" will be None
assert(img is not None)

# Display the image:
plt.imshow(img)
````

````python
img = cv2.imread('png_transparency_test.png', cv2.IMREAD_GRAYSCALE)

# If the image path is incorrect or there was another error, "img" will be None
assert(img is not None)

# Display the image:
plt.imshow(img)
````

````python
img = cv2.imread('png_transparency_test.png', cv2.IMREAD_UNCHANGED)

# If the image path is incorrect or there was another error, "img" will be None
assert(img is not None)

# Display the image:
plt.imshow(img)
````
